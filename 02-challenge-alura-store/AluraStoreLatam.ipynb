# Importación de datos

```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)
```

## Estructura de datos:
El conjunto de datos incluye la siguiente información:
- **Producto y Categoría:** Artículos vendidos y sus calificaciones.
- **Precio y Envío:** Valores de venta y costos asociados.
- **Fecha y ubicación de compra:** Información temporal y geográfica.
- **Evaluación de compra:** Comentarios de clientes.
- **Tipo de Pago y Cuotas:** Métodos utilizados por los clientes.
- **Coordenadas Geográficas:** Ubicación de las transacciones.

```python
# Leer y combinar los archivos
dataframes = [tienda1, tienda2, tienda3, tienda4]
df_total = pd.concat(dataframes, ignore_index=True)

# Verificar valores nulos
print("\nValores nulos por columna:")
print(df_total.isnull().sum())
```

# 1. Análisis de facturación

```python
# Calcular ingresos por tienda usando listas
ingresos_por_tienda = []

for i, df in enumerate(dataframes):
    precios = df["Precio"].tolist()  # convertir la columna a lista
    ingreso_total = sum(precios)     # sumar los valores
    ingresos_por_tienda.append([f"Tienda {i+1}", ingreso_total])  # guardar nombre y total

# Mostrar resultados
for tienda, ingreso in ingresos_por_tienda:
    print(f"{tienda}: {ingreso:,.0f}")
```

## Análisis de Ingresos Totales por Tienda
En esta primera fase del análisis, se calcularon los ingresos totales de cada tienda a partir de la suma de los valores registrados en la columna "Precio". Esta suma representa el total de ventas realizadas por cada establecimiento durante el periodo registrado.

Los resultados obtenidos fueron los siguientes:
- **Tienda 1:** S/. 1,150,880,400
- **Tienda 2:** S/. 1,116,343,500
- **Tienda 3:** S/. 1,098,019,600
- **Tienda 4:** S/. 1,038,375,700

Se observa que la Tienda 1 registra el mayor ingreso total, superando a las demás por un margen considerable. Le siguen las tiendas 2 y 3 con montos similares, mientras que la Tienda 4 presenta el menor ingreso acumulado. Estos datos permiten identificar diferencias en el rendimiento comercial entre tiendas, lo que puede deberse a factores como el tipo de productos vendidos, el volumen de ventas o la ubicación geográfica.

# 2. Ventas por categoría

```python
# Contar la cantidad de productos vendidos por categoría en cada tienda
ventas_por_categoria = []

for i, df in enumerate(dataframes):
    conteo = df['Categoría del Producto'].value_counts()  # cuenta ocurrencias por categoría
    ventas_por_categoria.append([f"Tienda {i+1}", conteo])

ventas_por_categoria
```

# 3. Calificación promedio de la tienda


```python
# Calcular la calificación promedio (satisfacción del cliente) por tienda
valoraciones_promedio = []

for i, df in enumerate(dataframes):
    promedio = df["Calificación"].mean()
    valoraciones_promedio.append([f"Tienda {i+1}", promedio])

valoraciones_promedio
```

# 4. Productos más y menos vendidos

```python
import matplotlib.pyplot as plt

# Crear una figura con 4 subgráficos (uno por tienda)
fig, axes = plt.subplots(4, 2, figsize=(14, 18))
fig.suptitle("Productos más y menos vendidos por tienda", fontsize=16)

for i, df in enumerate(dataframes):
    tienda = f"Tienda {i+1}"
    conteo_productos = df['Producto'].value_counts()

    # Productos más vendidos (top 5)
    top_productos = conteo_productos.head(5)
    # Productos menos vendidos (bottom 5, con al menos 1 venta)
    bottom_productos = conteo_productos[conteo_productos > 0].tail(5)

    # Gráfico de productos más vendidos
    axes[i, 0].barh(top_productos.index[::-1], top_productos.values[::-1], color='green')
    axes[i, 0].set_title(f"{tienda} - Más vendidos")
    axes[i, 0].set_xlabel("Cantidad de ventas")

    # Gráfico de productos menos vendidos
    axes[i, 1].barh(bottom_productos.index[::-1], bottom_productos.values[::-1], color='red')
    axes[i, 1].set_title(f"{tienda} - Menos vendidos")
    axes[i, 1].set_xlabel("Cantidad de ventas")

plt.tight_layout(rect=[0, 0, 1, 0.97])
plt.show()
```

## Análisis de Productos Más y Menos Vendidos por Tienda
Con base en los datos visualizados, se identificaron los productos con mayor y menor volumen de ventas en cada una de las tiendas. Este análisis es útil para comprender las preferencias de los clientes, ajustar el inventario y mejorar la estrategia comercial.
### Tienda 1
- **Más vendidos:** Los productos más vendidos incluyen tecnología y artículos del hogar, destacando el TV LED UHD 4K, el armario, y el microondas. Estos productos reflejan una demanda alta por equipamiento doméstico y entretenimiento.
- **Menos vendidos:** Entre los menos vendidos se encuentran artículos como Ciencia de datos con Python, pandareta y olla de presión, lo que sugiere un interés bajo en productos especializados o recreativos.
### Tienda 2
- **Más vendidos:** El liderazgo en ventas lo tienen productos educativos y musicales como Iniciando en programación y la guitarra acústica, además de artículos básicos como microondas y batería. Esto indica un balance entre formación, hobbies y necesidades domésticas.
- **Menos vendidos:** Aparecen productos como auriculares, sillón y mesa de comedor con bajo desempeño, posiblemente por saturación del mercado o falta de promoción.
### Tienda 3
- **Más vendidos:** Se destacan artículos de mobiliario y cocina, tales como kit de bancas, cama king y mesa de comedor. Esto refleja una demanda orientada al equipamiento del hogar.
- **Menos vendidos:** Entre los de menor rotación figuran guitarra eléctrica, mochila y bloques de construcción, lo que sugiere menor interés por productos lúdicos o personales en esta tienda.
### Tienda 4
- **Más vendidos:** Los artículos más vendidos incluyen cama box, cubiertería y mesa de noche, lo que indica una fuerte inclinación por artículos de dormitorio y cocina. También destacan Dashboards con Power BI, reflejando interés por herramientas digitales.
- **Menos vendidos:** En este grupo están lavadora de ropa, guitarra acústica, y ciencia de datos con Python, lo que podría evidenciar una falta de coincidencia entre la oferta y la demanda específica de esta tienda.

# 5. Envío promedio por tienda

# Visualización de Datos

```python
import matplotlib.pyplot as plt

# Nombres de las tiendas
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Cálculo de métricas
ingresos = [df["Precio"].sum() for df in dataframes]
valoraciones = [df["Calificación"].mean() for df in dataframes]
envios = [df["Costo de envío"].mean() for df in dataframes]

# Gráfico 1: Ingresos por tienda (Barras)
plt.figure(figsize=(10, 6))
plt.bar(nombres_tiendas, ingresos, color='skyblue')
plt.title("Ingresos totales por tienda")
plt.xlabel("Tienda")
plt.ylabel("Ingreso total (S/)")
plt.grid(axis='y')
plt.tight_layout()
plt.show()

# Gráfico 2: Valoración promedio por tienda (Línea)
plt.figure(figsize=(10, 6))
plt.plot(nombres_tiendas, valoraciones, marker='o', linestyle='-', color='green')
plt.title("Valoración promedio de clientes por tienda")
plt.xlabel("Tienda")
plt.ylabel("Valoración promedio")
plt.ylim(3.8, 4.1)
plt.grid(True)
plt.tight_layout()
plt.show()

# Gráfico 3: Costo de envío promedio por tienda (Dispersión)
plt.figure(figsize=(10, 6))
plt.scatter(nombres_tiendas, envios, color='red', s=100)
plt.title("Costo de envío promedio por tienda")
plt.xlabel("Tienda")
plt.ylabel("Costo promedio de envío (S/)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

# ¡Extra! Análisis del desempeño geográfico

```python
import seaborn as sns

# Asignar nombre de tienda a cada DataFrame
tienda1["Tienda"] = "Tienda 1"
tienda2["Tienda"] = "Tienda 2"
tienda3["Tienda"] = "Tienda 3"
tienda4["Tienda"] = "Tienda 4"

# Combinar todos los datos en un solo DataFrame
df_total = pd.concat([tienda1, tienda2, tienda3, tienda4], ignore_index=True)

# Gráfico de dispersión de las coordenadas geográficas por tienda
plt.figure(figsize=(10, 8))
for tienda, df in df_total.groupby("Tienda"):
    plt.scatter(df["lon"], df["lat"], label=tienda, alpha=0.6)

plt.title("Distribución geográfica de ventas por tienda")
plt.xlabel("Longitud")
plt.ylabel("Latitud")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Crear mapa de calor con Seaborn (heatmap de densidad de puntos)
plt.figure(figsize=(10, 8))
sns.kdeplot(
    data=df_total, x="lon", y="lat", hue="Tienda",
    fill=True, thresh=0.05, levels=100, alpha=0.6
)
plt.title("Mapa de calor de ventas por tienda")
plt.xlabel("Longitud")
plt.ylabel("Latitud")
plt.tight_layout()
plt.show()
```

El análisis de las coordenadas geográficas de las ventas permite identificar patrones espaciales relevantes sobre la distribución y concentración de las transacciones. A continuación, se describen los hallazgos por cada tienda:
